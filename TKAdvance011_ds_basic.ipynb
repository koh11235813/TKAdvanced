# ビッグデータ

今の社会で起きていることはデータ量の飛躍的な増大ですが、データは常に増え続けてきました。人が言葉を操れるようになった時、使えるデータ量はそれまでとは比べ物にならないぐらい大きくなりましたし、文字を書けるようになったとき、それを印刷できるようになった時と、いずれもデータ量は爆発的に増えました。近年では、ハードディスクが開発されたり、それが高度化したりするたびに、「これだけ記憶容量があれば、もうどんなものでも保存できる」と言われてきました。でも、私たちは容量があればあるだけ使い切ってきました。それまで文字データしか扱わなかったコンピュータに静止画や動画、音声を保存できるようになりましたし、今ではあまり見る気も無いテレビ番組まで全部録画しておいて後から思いついた時に見ることができる仕組みまで作られています。よく、データがたくさん発生することを「データ爆発」と表現しますが、データ爆発は常に起こり続けてきたと言えます。それでもなお、21世紀に入ってからデータの扱い方が劇的に変わったと言われており、「ビッグデータ」などの言葉が使われています。

### ビッグデータの定義

ビッグデータの正確な定義は研究者によって意見が分かれています。一般的によく言及されるのは、次の３点を満たすものです。

- 大量のデータ
- リアルタイムな発生
- 多種多様なデータ

大量という点は、イメージしやすいと思います。リアルタイムな発生はどうでしょうか。もともとデータというのは、とても貴重なものでした（今までももちろんそうですが）。したがって、どんな目的に使うのか熟慮した上で重複や矛盾が生じないようなデータの取得の仕方、保存の仕方を考え、取得した後にもよくチェックして活用していました。

少なくとも、今起こっていることをリアルタイムで知ることができるシステムやデータは一般的ではなかったのです。

でも、情報技術やデータの取得技法が向上し、素早く安価にデータが取れるようになりました。データを保存しておくためのストレージ機器も、高速で安価になりました。データはすぐに使えるものになったのです。

それに伴って、必要かどうかわからないけれども、とりあえずデータをとっておくことや、ある目的でとったデータを別の目的に転用することが簡単に行えるようになりました。例えば、防犯用にとっておいた監視カメラの映像を、人の動線の研究に使う例などが典型的です。こうした背景があって、わたしたちが取り扱えるデータは極めて**多様**になりました。データの種類という意味でもそうですし、データを測定する機器の意味でもそうです。文字だけだったデータは、映像や音声を含むようになり、それらを相互に変換する技術（音声を文字起こしするなど）も発展しました。データを測定する機器も、研究室に置いてあるような大型の測定器ばかりでなく、軽くて薄くて安価なものが街中に溢れるようになりました。スマホには温度センサー、位置情報センサー、ジャイロセンサー、音声センサー、画像センサーが満載され、店舗に行けば**RFID**(電子タグ）が商品に貼られ、商品の管理にも万引き防止にも使われています。

３つの定義の中には明記されていませんが、ビッグデータという言葉の中には、「みんなが使える」、「必ずしもかっちりしたものでなくていい（非構造化データ）という雰囲気が含まれていることがあります。

### IoTとビッグデータ

ビッグデータと密接な関わりを持ちながら社会に根付いてきたものが、**IoT**です。IoTは**Internet of Things**、日本語ではよく「モノのインターネット」と紹介されます。インターネットなんて元々モノとモノ（コンピュータとコンピュータ）をつないでいるじゃないか、と思うかもしれませんが、少なくとも今までのインターネットはコンピュータの背後に人間がいることを想定していました。

WWWであれば、Webサーバから送られてくるWebページのデータは私たちのWebクライアント（ブラウザ）に届きますが、その背後にはWebページを見ようとしている利用者がいるわけです。

これがIoTになると、サーバとセンサーのように、**人間を介さない接続**が一般化します。人間のオペレータが寝ている間も、センサーは黙々とサーバに情報を蓄積し続けるわけです。さらに、当初は想定していなかったセンサーからも、サーバは情報を受け取るようになるかもしれません。

従来は、「このセンサーから情報を取得する」とあらかじめ決めておき、それ以外からの情報は得られなかったのですが、IoT時代には「こういう情報をこれくらいの品質で、こんなデータ形式で公開しているセンサーがあればそこから情報を得てくる」といった「条件」を決める形になるでしょう。条件に合致するセンサーをサーバーが勝手に見つけて、勝手にデータを取得してくるイメージです。人手を介する部分が確実に減っていることがわかると思います。

### 5GはIoTを促進

さらに**5G**(**第5世代移動通信システム**）が加わります。5Gの特徴は**高速大容量**、**低遅延**、**多端未接続**です。このうち、多端未接続はIoTを促進することに目標があると言われています。今まで移動通信システムはスマホとガラゲーのためのものでしたが、それだけではなくて、IoTで使われるセンサーを結ぶネットワークとして使われるということです。5Gには速度も遅くて、通信遅延も大きくていいから、大量のセンサー類を安く、バッテリーを節約して使うモードが設定されています。

これによって今以上に大量のセンサーを地球上にばら撒くことができます。センサーを設置して、データを取得するとき、問題になるのはその取得方法です。Wi-Fiでは接続できる距離は短いですし、有線ケーブルだと屋外に設置する障害になります。5G回線によって、今まで難しかった山地や公海にもセンサーを置くことが容易になるでしょう。これにより、気象予測がもっと正確になるかもしれませんし、農地に高密度なセンサー群を配置して、農業を自動化するような構想も練られています。

あらゆる場所に多様なセンサーを設置して、私たちの社会や環境をより深く知ろうとする試みのことをことを**トリリオンセンサー**と言います。トリリオンは１兆のことで、年間１兆個のセンサーを設置していこうとするものです。

# 進化するテクノロジーと社会

ビッグデータやIoT、5Gは技術の用語でしたが、これらを駆使することで私たちの社会がどう変わっていくのかを表す言葉に、**第４次産業革命**やS**ociety5.0**があります。社会と技術は互いに刺激合う関係にあります。例えば、情報技術が洗練されたことと、音楽のライブ化、握手会の隆盛などは関係があります。無償に近いコストでほぼ劣化のないコピーを作ることができるようになり、楽曲の売り上げは減少しました。違法コピーをなくすことはもちろん必要ですが、なかなか根絶できるものではないので、アーティストたちはライブや握手会に力点を置くようになりました。これらの「体験」をコピーするのはまだ難しく、オリジナルであれば人はお金を払ってくれるからです。「体験」をコピーする技術として**VR**がありますが、まだ「VRで握手したから、本物とは握手しなくていい」ほどにはなっていません。これはほんの１例で、技術が進化すると社会は変わりますし、その逆も言えます。第４次産業革命はそれを丸っと言い表した言葉です。

### 第４次産業革命

第１次産業革命が蒸気機関、工場制機械工業、第2次産業革命が電話や電気、印刷機、内燃機関、第3次産業革命がコンピュータやOA、FAに象徴されるとしたら、第４次産業革命の象徴はAI、ロボット、ナノテクノロジー、バイオテクノロジーなどです。

古くは機械工学やシステム工学を融合させたサイバネティクスという言葉がありました。サイバーといえば、今は仮想空間やインターネットを表す言葉になっていますが、仮想と現実のオーバーラップも、第４次産業革命の重要なテーマになるでしょう。私たちは既に、車のフロントガラスなどに風景と速度表示や注意表示が重なって見える表現に慣れています。仮想空間で稼いだお金やアイテムを、現実の貨幣に換金できるケースも増えてきました。仮想と現実は互いに徐々に侵食しつつあると言えます。

第４次産業革命では、単純作業のような仕事はもはやAIに取って代わられるようになるとか、反対に高度なICTやAIを駆使する分野では大変な人手不足が起こるなどと言われています。この新しい社会に対応するために、今まさに私たちはAIやデータサイエンスのリテラシーを学んでいるわけです。

### Society5.0

Socieity5.0は、日本が世界に先んじて示した新しい社会のありようで、仮想空間（サイバー空間）と現実空間（フィジカル空間）の融合がキーワードです。多くの点で第４次産業革命と重なる考え方ですが、第４次産業革命がまさに産業に軸足を置いているのに対して、Society5.0はそこで生まれた技術をもとに社会のあり方を変えることを主眼としています。

Society1.0が狩猟社会、Society2.0が農耕社会、Society3.0が工業社会、Society4.0を情報社会として、その次に来る社会のイメージです。多分に理念的なものなので実感を伴うのはまだこれからですが、もしキャプションをつけるとしたらサイバー社会でしょうか。生活の中にAIやロボット、センサー、ウェアラブルデバイスなどが無理なく溶け込んで溶け込んで協働している社会です。

### クラウドとエッジ

クラウド・コンピューティングによって膨大な計算をクラウドで計算できます。一方、エッジ・コンピューティングはユーザーに近い端末で計算を行うため、計算遅延が起こらないこと、セキュリティーやプライバシーが守られるメリットがあります。

エッジコンピューティングのような形でAIに学習をさせる連合学習（Federated Learning）という方法も考案されています。利用者のパソコンやスマートフォンなどの端末に機械学習のモデルをダウンロードし、端末に入っているデータを使って機械学習を行います。データは送らずに、変更点のみをクラウドのサーバーに送信すると、他の端末からの結果も併せて機械学習のモデルを改善することができます。

# データ駆動型社会

仮想空間と現実空間が結びつき、IoTや5Gで多種多様なデータが取れるようになると、データとそこから導かれる知見が大きな価値を持つようになります。昔からデータには**価値**がありました。朝焼けは雨、夕焼けは晴れといったことわざは、多くの人が集めたデータと知恵で未来予測をしようとした、**人力ビッグデータ**の試みと言えます。しかし、第４次産業革命やSociety5.0で用いられるデータの量と種類、それを処理する速度は、この頃とは隔世の感があります。天気予報の精度はとても向上しましたし、例えばスポーツなど、他分野への応用も盛んです。

### データ駆動型社会とは

私が子供の頃は、野球が下手だったらとにかくゴロを転がせと教わりました。でも、データをきちんと取ってみると、フライを打ち上げた方が出塁や打点につながることがわかってきました。それまで常識だと考えられてきたことが、データの裏付けによってそうでもないことが明らかになってきたのです。

物事の予測精度も上がりました。選挙の予測システムは当選者をかなり正確に言い当てるようになりましたし、自動車レースのF1では事前予測があまりに正しいので、レースを見るのがつまらなくなったと言われるほどです。

データの価値が増大し、データによって世の中が動く様子を、**データ駆動社会**と言います。自動車のセンサーは高度化し、車が壊れる前に予防修理を促されたり、ウェアラブルデバイスで体のバイタルサインを常に取ることで、病気の予兆を捉えようとする仕組みなどもかなり普及しました。チケットの抽選では、誰は落としてもまた参加してくれるとか、誰を落選させるともう観に来る気がなくなってしまうのか、などといったことも調査されています。

高齢化社会に伴って、次のトイレのタイミングを教えてくれるセンサーや、飲み忘れがないように胃でとける時に電気信号を出す薬などが研究されています。地震の揺れが自分のところへ来る前に警報を発するシステムなどは一種の未来予測ですが、経験された方も多いのではないでしょうか。

スポーツで思いもよらない戦い方を発見したり、実際に揺れる前に地震が来ることがわかったりすれば、それでゲームに勝ったり、命を救ったりすることができるかもしれません。データに大きな価値があると言われるのはそのためです。何か話をすると「その話にエビデンス（証拠となるデータ）があるのか」と言われたりもします。私たちがデータサイエンスを学ぶ理由はそこにあります。

# 知的財産権と個人情報の保護

個人の詳細なデータがあれば、カスタマイズしたサービスが提供できるかもしれません。しかしそれは、個人の秘密や私生活（プライバシー）を侵害することに繋がりかねません。AIシステムや制度を用いてプライバシーや個人情報を保護する方法が考案されています。

### 著作権法

**著作権法**は創作された表現物を保護するための法律です。著作権は特許権と異なり、創作された段階で発生します。本来は小説や音楽を対象とした概念でしたが、改正によりプログラムやデータベースなどもその保護対象に含まれることが確認されています。開発言語やアルゴリズムなどはその対象に含まれないことと、会社の業務で作成した著作物は特別な契約がない限り、会社に著作権が帰属する点に注意が必要です。著作権は「**著作者人格権**」「**著作財産権**」「**著作隣接権**」によって構成されます。著作者人格権が譲渡不能な権利であるのに対して、複製権、公衆送信権といった著作財産権は売買が可能です。なお、著作権の有効期限は70年です。

データセットの利用に関しては、日本の著作権法は非常に柔軟な条文が盛り込まれており、AI開発を目的とする場合、一定条件のもとで著作権者の許可なく著作物を利用できるようになっています。これは元々**著作権法４７条の７**によって実現されており、とりわけ機械学習をする際は多大な恩恵を受けてきました。しかし、著作権法４７条の７は、あくまで「データ収集からAI生成までの一連の流れ」の中でしか適用されないという制約があったため、AI生成を行うために用意した学習用のデータセットを著作権者の許諾なく第三者に販売したり、共有したりする場合には適用されず、著作権侵害となりました。

それが2019年1月1日から施行された改正著作権法により拡張され、著作権法４７条の７に代わる著作権法３０条の４が規定されました。**著作権法３０条の４**によりAI生成のために用意した学習用データセットを不特定多数の第三者に販売したり、公開する事が適法となりました。

### 個人情報保護法

個人情報保護法とは、事業者を対象として個人情報の取り扱いに関するルールを定めた法律である。個人情報は、個人情報保護法上、次の４つに分類される

1. 個人情報
2. 個人データ
3. 保有個人データ
4. 要配慮個人情報

##### 個人情報

個人情報とは、

1. 生存する個人に関する情報であって、名前、生年月日、住所など特定の個人を識別できるもの
2. 個人識別符号を含むもの

を意味します。

個人識別符号とは、身体の一部を変換した符号、またはカード等の書類に記載された符号のいずれかに該当するものを指す。例えば、顔、指紋、声紋、DNA、歩行の態様、マイナンバー、パスポート番号などが挙げられる。画像から個人の目、鼻、口の位置関係等の特徴を抽出し、数値化するデータ（特徴量データ）も個人識別符号にあたる。



##### 個人データ

個人データとは、特定の個人を検索することができるように構築したデータベースに含まれる個人情報のことをいう。このデータベースを個人情報保護上、個人情報データベースといい、個人情報データベースを事業のために使用しているものを個人情報取扱事業者という。個人情報取扱事業者には、個人データが漏洩、滅失しないよう適切な安全管理措置をとることが求められたり、個人データの第三者提供の制限等を受けたりというように、一つ目の個人情報を取り扱う事業者よりも厳しい義務が課される。

##### 保有個人データ

保有個人データは、個人データのうち、個人情報取扱事業者に開示・訂正・消去等の権限があり、６ヶ月を超えて保有するものである。保有個人データを持つ個人取扱事業者は、個人データを取り扱う事業者に課される義務に加え、保有個人データに関する事項の通知や保有個人データの開示、訂正、利用停止などの義務が課される。

#### 要配慮個人情報

要配慮個人情報とは、人種、信条、社会的身分、病歴、犯罪の経歴、犯罪被害を受けた事実、その他本人に対する不当な差別、偏見などが含まれた個人情報のことである。個人情報の中でも特に慎重な取り扱いが求められる情報であり、2017年5月の法改正で明確化された。要配慮個人情報にあたる情報は、取得に際して本人から同意を得る必要がある。

### 匿名加工情報

2017年に施行された改正個人情報保護法では「匿名加工情報」という制度が導入されました。個人を識別できないように情報を工夫することで、個人の情報の利活用を促進すると同時に、個人情報を保護することが可能になります。

プライバシーの権利とは、昔は自分の秘密を暴かれない「放っておいてもらう権利」でしたが、現在では、「自分の情報を自分で管理しコントロールする権利」と言われています。コントロールの一環としては、近年ではさらにインターネット上から、自分にとって都合の悪い情報を消してもらう権利（消去権）も重要視されています。

# セキュリティと安全

サイバー空間上はさまざまな攻撃にさらされています。AIもネットワークに繋がった時、情報を不正に入手しようとする攻撃などにさらされます。そのため、AIサービスやシステムにもセキュリティ対策を考えなくてはなりませんんが、技術だけでなく組織的、そして人の意識改革が必要となってきます。

### 情報のCIAとは

情報セキュリティは、**情報の機密性**(Confidentiality）、**完全性**（Integrity)、**可用性**(Availability）を維持することと定義されています。これらの頭文字をとって、「情報のCIA」と呼びます。

情報の機密性とは、その情報に許可されたものだけがアクセスできることを示す用語です。アクセスコントロールとも呼ばれます。情報の完全性とは、その情報が不正に書き換えられたり、消されたりしていないことです。また情報の可用性とは、ユーザーが情報が必要な時に、いつでも使える状態にあることを言います。情報を提供するサーバ機器が故障していたりすることで可用性は低下します。

### 誰がなぜ攻撃してくるのか

サイバー攻撃の目的や理由は、興味本位や腕試し目的の愉快犯や金銭目的、さらには国家的•軍事目的などさまざまです。愉快犯や金銭目的であれば、防御を堅牢にすることで、攻撃のコストが見返りに見合わなくなるので攻撃を避けられるかもしれません。しかし、国家や軍事目的の場合、コストを度外視しても情報を入手したい、攻撃したいということもあるかもしれません。近年はウイルス感染などの攻撃だけでなく、フェイクニュースを流すこと自体もサイバー攻撃と見做されます。

### サイバー攻撃のアプローチの多様性

#####特定のターゲットを狙ったサイバー攻撃


*   **ランサムウェア** : 「ランサムウェア」はマルウェアの一種で、端末をロックしたりファイルを暗号化したりして使用不能にしたうえで、身代金（ランサム）を要求するものです。

*   **標的型攻撃による機密情報の窃取** : 
特定の組織や個人の機密情報等を狙い、PCをウィルスに感染させて組織内部に潜入するものです。手段としてメールの添付ファイル、不正アクセスなどが使われます。

*   **詐欺メールによる金銭の窃取** :
虚偽の内容のメールを使用して金銭を騙し取るものです。組織では取引先を装ったビジネスメール詐欺、個人では架空請求や知り合いを装った寸借詐欺などがあります。



##### 不特定多数を狙ったサイバー攻撃

 * **フィッシング詐欺** : 
官公庁や金融機関など、実在の組織を騙ったサイトを設け、入力フォームから氏名や生年月日、クレジットカード番号、ログイン情報などを抜き取るものです。

* **SMS（Short Message Service）を通じたマルウェア感染** : 
宅配業者などの名を騙ったSMSをスマートフォン宛に送信する手口です。SMSに記載されたURLをクリックすると、自動的にマルウェアがインストールされてしまいます。

* **偽の警告画面による金銭の窃取** : 
「このPCはウィルスに感染しています」などのメッセージを画面に表示し、電話等で連絡を取らせて金銭を騙し取るものです。

##### サーバーに負荷をかけるサイバー攻撃


* **DDoS攻撃** : サービス妨害（Denial of Servise: DoS）は、サーバに負荷を集中させるなどしてサーバを使用不能に陥れる攻撃です。1台のPCで攻撃を行うDoS攻撃、多数のPCを踏み台として攻撃を行うDDoS攻撃があります。

* **SYNフラッド攻撃** : 
ウェブサーバーに接続要求（SYN）を大量に送り付け、サービス停止に追い込むものです。切断要求（FIN）を用いたFINフラッド攻撃もあります。

* **UDPフラッド攻撃** : 
IPアドレスを偽って攻撃を行います。UDPという通信方式の「通信相手の応答を待たずにデータを送り続ける」という特性を利用して大量のデータを一方的に送信し、標的サーバーを停止させます。

* **ブラウザの再読み込みによる攻撃** : 
ウェブブラウザの再読み込み機能を利用してWebサーバーに大量の要求を送り付けるものです。更新には主にF5ボタンを使うことから、俗に「F5アタック」とも呼ばれます。

##### OS/Webサイトなど脆弱性を狙ったサイバー攻撃


* **SQLインジェクション** : 
データベースへの命令文である「SQL」を使用し、Webアプリケーションの脆弱性を突いてデータの消去や改ざん、情報漏えい狙うものです。

* **OSコマンドインジェクション** : 
ウェブサーバーへのリクエストにOSへの命令文を紛れ込ませ、不正に実行させるものです。サーバー内ファイルを改ざん、削除、流出される恐れがあります。

* **クロスサイトスクリプティング** : 
Webページを動的に生成するWebアプリケーションの脆弱性を突いて任意のスクリプトを実行するものです。クッキーの漏出や個人情報の漏えいの恐れがあります。

* **ゼロデイ攻撃** : ソースコードなどが無償公開されており、改良や再配布が自由にできるオープンソースウェア（OSS）は無償で使える代わりに、利用者は自己責任で使う必要があります。そのためOSSにセキュリティ上の問題や脆弱性が発見されたときは、利用者が対応をしなければなりません。直ちに利用者が対応を取らなければ、ハッカーなどに攻撃されてしまいます。脆弱性情報が公開されて対応するまでの間に行われた攻撃のことをゼロデイ攻撃といいます。



##### AIを狙ったサイバー攻撃

AIは現在、自動運転や医療機器などさまざまなところに使われるようになっています。また、サイバーとフィジカルが融合した空間では、物理空間上で動くAIに対してもサイバーから攻撃が繰り返されることを注意しなければいけません。例えば、自動運転車が制御できなくなったり、真夏に空調が切られたり、全てが青信号になってしまったりすると、私たちの生活の安全が脅かされます。AIが私たちの生活に身近になる程、サイバー攻撃の被害は大きくなります。

具体的にはどのように攻撃をするのでしょうか。意外なものを使ってハッキングが可能な事例が報告されています。2019年にスマートスピーカーなどに対して、レーザー光線をマイク部分に照射することで、操作が可能という報告がなされました。100mも遠く離れた場所から、望遠レンズを使ってレーザーをマイク部分に当て、電気信号を生成させ、AIが音声を拾ったかのように錯覚させました。これによって、家のガレージのドアを開けることができたのです。これをライトによる命令（[Light Commands](https://lightcommands.com/index_jp.html)）と言います。

##### 内部の人間を狙ったサイバー攻撃

サイバー攻撃に対する技術的な備えは重要です。しかし、セキュリティを破ろうとする人は、愉快犯や腕試しをしようとしているのでない限り、技術的に真正面から攻撃してくるわけではないことに注意が必要です。

* **ソーシャルフィッシング** : 人に対して言葉巧みに情報を漏らさせるような会話術を使った攻撃です。技術的にどんなに防御が完璧でも、守る人たちの意識が低ければ意味がありません。サイバー空間上は万全であっても、物理空間上での防御が疎かになっていては本末転倒ですが、人の心理をついたさまざまな攻撃が考えられます。コーヒーなどを持って勝って知ったる感じの人が「中にいる人に持ってきてくれと頼まれた」と言ってにっこり笑ってくるので、うっかり部屋に入れてしまった、など大きな組織や出入りが激しい組織で使われる手口かもしれません。

* **スタックスネット** : 「重要なデータはインターネットに接続されていないから、情報漏洩やハッキングのリスクはない」と思っていても油断はできません。2010年ごろ、イランの核燃料施設の制御システムを乗っ取ったスタックスネットというマルウェアがあります。このマルウェアはUSBメモリに入っていて、関係者が制御システムにUSBを差したことで感染しました。このUSBは技術者が集まる会議のお土産として無料配布されたグッズに忍び込まされていたと考えられています。このように内部の人の手を介して、攻撃的なマルウェアが入り込んでしまう手口もあるのです。

* **セキュリティ対策が甘い下請け企業を狙った攻撃** : 情報を入手するためならば、セキュリティー対策が万全な企業ではなく、セキュリティーが甘そうな下請け企業などを狙う作戦もあります（[デンソーの子会社の例](https://www.itmedia.co.jp/news/articles/2203/14/news114.html)）。中小企業が多く、サプライチェーンが長い日本では、大企業並みに中小企業がセキュリティにコストをかけられるかといった課題をついた攻撃です。

### セキュリティ対策

今までのセキュリティ対策は、企業や組織のネットワークの内側と外側を区別して、外側からの不正な侵入を防ぎ、内部にいる人は信用するという考えでした。しかし、情報技術を取り巻く環境が複雑になっている現在、このような従来の考え方では攻撃を防ぎきれなくなっており、新たな対策が求められています。

### AIによる対抗手段

　AIがサイバー攻撃のパターンを学習して対抗すると言った研究も行われています。マルウェアの種類など、過去の類似の事例からその傾向と対策をAIが考え出すというものです。

### ゼロトラスト

オープンソースソフトウェアを用いるようになったり、COVID-19の関係で在宅勤務が増えるなど、現在、組織の枠を超えて、人やデータ、ソフトウェアが行き来するような状況になりつつあります。そこで、ネットワークの内外という基準で信頼する/しないを分けない、つまり基本的に**ゼロトラスト（信頼しない）**という考え方が注目されています。

ゼロトラストという考え方でセキュリティ対策を行うと、「どこからのアクセスでも認証が必要になる」「最低限の権限しか利用者には与えない」「情報やネットワークに関係する通信を全て自動で監視（モニタリング）する」などのルールや仕組みを構築することが必要になります。

### 公開鍵暗号、ディジタル署名、PKI

通常の暗号（**共通鍵暗号**）は暗号化の鍵と復号（元に戻すこと）の鍵が同じです。暗号化した人は、、復号してもらいたい人に安全に鍵を届けるのに苦労します。よく「パスワードは次のメールで送ります」とか「パスワードは私の内線番号です」のような伝え方をすることがありますが、セキュリティ的には無意味です。なぜなら、前のメールの内容が漏れているなら、次のメールの内容も漏れているはずですし、数桁の内線番号程度のパスワードは、総当たり法で、一瞬で破られてしまうからです。

これに対して、**公開鍵暗号**は、暗号化の鍵と復号化の鍵が異なります。情報の受け手（Aさん）は、暗号化の鍵と復号の鍵のペアを生成し、暗号化の鍵だけを公開し（公開鍵）、復号の鍵は秘密にしておきます（秘密鍵）。Aさんにメッセージを送りたいBさんは、Aさんの公開鍵を用いてメッセージを暗号化して送ります。一方、Aさんは自身の秘密鍵で暗号化されたメッセージを復号し、メッセージを読むことができます。メッセージは秘密鍵を持つAさんしか読むことができません。

公開鍵暗号の応用として、作成した文書にディジタル署名を付けることができます。**ディジタル署名**は「**なりすまし**」および「**改ざん**」を防ぐ技術です。ここでは、AさんがBさんに契約書を送る場合を例にディジタル署名の手順を説明します。契約書にディジタル署名を添付することで、送信者がAさん本人であることと、契約書の内容に改ざんがないことを示せます。

1.   Aさんは秘密鍵と公開鍵のペアを作り、秘密鍵をAさんがもち、一方を公開鍵として、あらかじめBさんに送ります。
2.   Aさんは契約書の本文からその契約書に固有の値である**ハッシュ値**、または**メッセージダイジェスト**を求めます。ハッシュ値を求める方法はBさんも知っているとします。
3.  Aさんは、手順２で作成したハッシュ値を秘密鍵で暗号化します。これがディジタル署名となります。Aさんは契約書の本文とディジタル署名を、ネットワークでBさんに送ります。
4. Bさんは、受け取った契約書の本文から、ハッシュ値を求めます。仮に123456となったとしましょう。さらにディジタル署名をAさんの公開鍵で復号します。仮に、これも123456となったとします。両者が一致したので、契約書はAさんが送ったものであること（なりすましがないこと）と、改ざんがないことが確認できました。なぜなら、ディジタル署名をAさんの公開鍵で復号できたのは、ディジタル署名がAさんの秘密鍵で暗号化されていたからです。また、契約書の本文から求めたハッシュ値と、復号したハッシュ値が一致したのは契約書の内容に改ざんがなかったからです。


公開鍵暗号の仕組みを公的な場で使うには、公開鍵と本人とを結びつける仕組みが必要です。そのための仕組みが**公開鍵基盤**（**PKI**)です。

PKIの大規模な応用例として、マイナンバーカードのICチップには、本人の公開鍵・秘密鍵と、その公開鍵が本人のものであることを証明するための電子証明書が入っており、マイナンバーカードをカードリーダーやスマホにかざすだけで、本人であるという認証ができたり、書類に署名したりできる**公的個人認証(JPKI）**という仕組みが作られています。この仕組みを活用できれば、役所に行かなくても全ての手続きがオンラインで完結する**電子政府**が完成します。
