# **1. AIの歴史**

## **AIとは**

AIとは、Artificial Intelligenceの略語です。日本語に訳すときは、よく人工知能という言葉を使います。いまやAIという言葉を使わない日はないくらい生活の中に浸透してきていますが、実は気軽に使わないほうがいい言葉かもしれません。

というのも、初期の研究者が考えていたAIは、「人工の知能」すなわち人間の代わりになるようなものを指し、それを研究していたからです。人間のように筋道を立てて考え、ひょっとしたら感情を持ったり理解したりできることも期待されていました。

研究が始まった頃に比べると、コンピューターを専門家のように振る舞わせる技術はとても進歩しましたが、人の変わりができるようになったわけではありません。そのため、研究者によっては未だAIは実現しておらず、そんな言葉を使ってはいけないという人もいます。

### 強いAIと弱いAI

一方で、特定の用途に限って言えば、もう人間と同等だったり、人間を上回る能力を持つシステムも現れてきました。オセロや将棋、囲碁で人とシステムが勝負して、システムの方が勝ってしまうことはもう普通になりました。だから、これをAIと読んで差し支えないという人もいます。

どちらも一理ありますが、ややこしくなったり、混同したりすると良くありません。そこで、両方を区別する必要があるときには、人の代わりになるような、例えば特定の分野だけでなくあらゆる分野で人と同じように活動して、感情すら理解できるようなAIのことを「**強いAI**」、そうでなくてオセロならオセロ、将棋なら将棋、自動運転なら自動運転だけをすることができるAIのことを「**弱いAI**」と読んで区別することがあります。今、報道で目にしたり、商品として触れる機会があるAIは「弱いAI」ですが、テキストなどを読むときはどちらの意味で使われているのか、文脈に注意するようにしましょう。

## **AIブーム**

技術の普及はそうそううまくいくものではありません。AIブームも実は今回が３回目なのです。１回目、２回目のブームも期待が高まりましたが、幻滅期を越えられずに退潮して行きました。

第一次AIブームは1960年台に起こり、1956年の**ダートマス会議**において、コンピュータを使って人間のような知性を再現する研究の名称として「人工知能」という用語が生まれ、広くその言葉が普及しました。この第1次AIブームの中心にあった手法は**探索**と**推論**です。この二つの手法に関するアルゴリズムの発展により、シャノンのチェスプログラムやハーバート・サイモンとアレン・ニューウェルらによる「ロジック・セオリスト」といった当時としては画期的なプログラムが発表されました。ロジックセオリストは、「数学原理」に掲載されている幾つかの定理を自動で証明するというものです。また、今でいうニューラルネットワークの基礎が築かれるなど、新しい技術の発展にわきました。すぐにでも人を超える知能が生まれるとまで言われましたが、後で述べる**フレーム問題**などの壁にぶち当たってブームが去りました。人工の知能を作るのはとてつもなく大変そうだと、みんなが気づいた時期です。

第二次AIブームは1980年代で、**エキスパートシステム**が提唱されました。専門家の知識を**プロダクションルール**化（if - then ルール）していって、「この症状が出たら、この病気」などとお医者さんのように振る舞ったり、「この局面で株を買うと大儲けできる」とトレーダーとして使えるシステムなどが考案されました。「人間を完全に置き換える」ようなものを夢想しない分、みんな冷静になったのだと思います。目標が「○○の専門家」に小さくなっていますから。しかしながら、それでも幻滅期が訪れました。専門家の肩代わりができるくらいの知識を覚え込ませるのは、本当に大変なのです。例えば将棋で、「飛車と桂馬だったら、明らかに飛車のほうが価値が高い」、「でもいまは桂馬があると相手の王様が詰むので、桂馬の価値は無限大だ」などといちいちやっていく面倒さを考えてみてください。すぐに何億通りものルールになってしまいます。

この第二次AIブームの時代、日本でも**第５世代コンピュータ**という非常に大きな官民を挙げた国家プロジェクトが動いていました。通商産業省(現在の経済産業省）が中心となり、当時５７０億円という巨額の予算が投じられ、従来のコンピュータとは異なる非ノイマン型の人工知能型コンピュータを実現しようとする目標のもと、日本中のAI研究者、関連企業を巻き込み進められました。このプロジェクトでは、当時、世界最速の推論コンピュータを実現するなどの成果は上げられたものの、実社会への応用につながる成果はほとんど出すことができずに終わりました。

今の第三次AIブームは、基本的にはエキスパートシステムの流れを受け継いでいます。いま、AIと言われているものは、自動運転の専門システムだったり、チェスの専門システムであったり、〇〇の専門家という意味ではエキスパートシステムそのものです。

エキスパートシステムはルールを覚えさせるのが大変で失速しましたが、機械学習による学習の自動化や、多層のニューラルネットワークモデルを利用した**深層学習（ディープラーニング）**の登場により、楽にすることができました。これでAIシステムづくりに必要な労力が劇的に改善したわけです。深層学習の最大の特徴は、その高い学習性能とデータから直接、**特徴量**を計算できる点にあります。データの特徴量とはデータの特徴を定量的に表現した値であり、従来の機械学習では人間が学習に必要な特徴量を試行錯誤的に定義する必要がありました。一方、深層学習にも弱点があります。それは、質の良い学習器を作成するには、大量の質の良いデータと膨大な計算資源を必要とするということです。深層学習は、インターネットやIoTが普及したことによるビッグデータ取得の容易化、GPUを含めた計算機性能の向上という現在の状況だからこそ生まれた技術と言えるのかもしれません。

第三次AIブームの特徴は、AI技術が幅広い製品やサービスに活用され、これまでのブームにない規模で実社会に影響を及ぼしている点です。

## **フレーム問題**

社会の隅々まで「AI的」サービスが浸透し、使われ始めているのは皆さんご存知の通りです。ただ、いまブームの渦中にある「AI」が万能でないことは、十分理解しておいてください。**フレーム問題**など、いくつかの難しい問題が今も未解決なままなのです。

フレーム問題とは、起こりうる問題すべてに対処できないことを指す用語です。例えば、「お使いに行って豆腐を買ってきて」は比較的簡単なタスクと思えます。状況によっては未就学児でも実現できるでしょう。しかし、現実にこのタスクをAIを搭載したロボットにやらせるには、車を避けられるか、雨が降ってきたらどうするか、通りすがりの人に抱きつかれたら？突然隕石が降ってきたら？など無数の事象に対応しなければなりません。この「無数の事象への対応」を現実的な時間内で解くことは、今のAIにはまだできないのです。

もちろん、こうした事象のほとんどは、まず遭遇することがありません。私たちも意識していません。そこが人間の優れているところで、ほとんど起こりそうもないことは意識から除外して行動することができます。AIにもそれを試みさせようとしていますが、何が重要で何が重要でないかの判断はとても難しいのです。将棋AIでも「よさそうな手」の先だけを選択して読むのが難しく、結局はルール上指すことができるすべての手を読んでいるAIが多いです。人間の騎士は手を読む速度はAIに敵いませんが、「よさそうな手」だけを選んで考える能力は圧倒的です。将棋であれば、ありそうな手を手当たり次第に読むことはまだ現実味がありますが、現実世界のお使いなどでは困難です。いくらでも起こりうる例外的な事象に際して、どう回避するか、あるいは目的を再設定したほうがいいのかは、単にお使いのことだけ考えれば良いわけでなく、（事故に遭いそうだったら）命の価値なども含めた判断になるわけです。これを全く無視して仕舞えば、お使いのために人命を軽視する危険なAIさえ出来上がりかねません。「特定の範囲（フレーム）を越えた事態に対処すること」は、現時点ではAIには荷が重いと言えます。

# **2. AIと社会**

## **説明可能性と透明性**

深層学習では、ニューラルネットワークモデルが多層であるため、評価や判断の内容を明確に説明できないことがあります。これがいわゆる「ブラックボックス問題」です。 ブラックボックス問題に対抗するため、ブラックボックスの中で何が起きているかを画像や言葉で説明する技術として、**説明可能AI**(Explainable Al:XAI) の開発も進められています。

### 説明可能性とコストのトレードオフ


ブラックボックスになってしまう AIシステムに対して、説明可能性が求められていま す。しかし、説明可能なAIの構築には、コストがかかる場合があります。医療や人事採用の判断など、人の命や将来にかかわる重要な場面では説明可能性が求められるものもあるかもしれません。しかし、AIの予測や提案を人間が参考程度にしか使っていない場合、説明可能なAIを構築することにお金をかけるよりも、AI判断の精度を上げることに費用をつぎ込んだほうがいいこともあるでしょう。目的と利用できる資源、利用者の要望などを考慮してAIシステムを構築する必要があります。


###求められる透明性

AIがどのように動作するかを技術的に説明するだけでは、人々の不安が払しょくされずに、AI システムの利用が進まない場合があります。
重要なのは、何かAI システムに問題が起きたときに、誰が責任を取るのか、補償はあるのか、人と機械の役割分担や責任の所在が はっきりしていることです。
そのためには、AIシステムが説明可能であるという技術的な透明性だけではなく、責任を誰が取るかという法的あるいは制度的な透明性も求められています。透明性を確保することによって、AI システムに対する信頼も確保できるようになります。

###説明可能であれば攻略される

機械であれ人間であれ、判断の基準が数値化されると、対策が立てられて攻略される可能性があります。
たとえば、人事採用AIにおいて「ある単語が 入っていると、高評価を下す」などがわかる と、エントリーシートにはその単語を含めた 記述が増えるでしょう。
人間であれば、あからさまに対策を立てられたら、臨機応変に基準を変えたり問題設定を変えたりできます。しかし機械的に判断するAIシステムに、そのような柔軟性はありません。そのため、重要な場面では人間の最終判断が求められます。

##**AIの公平性**

機械学習は統計学を基礎としているため、過去と未来は変わらないという前提で認識や予測を行います。つまり、現在の社会に存在する差別や偏見を再生産してしまうのです。過去から現在を形作るのではなく、 どのような未来を作っていきたいのかを考えるのは人間の役割です。

### アルゴリズムの調整



社会にすでに普及しているデータだけではなく、社会構造そのものに偏りがある場合も
**アルゴリズムの調整問題**が生じます。たとえば、テクノロジー業界で働いている女性の数は世界的にも少ないです。また、工学系の有名大学に入学する人は、裕福な家かつ都市部出身者が多いと言われており、このようなテクノロジー業界の多様性のなさも、無意識のバイアスの形成につながると指摘されています。



無意識のバイアスに気がついたとして、どのようにデータやアルゴリズムを調整すればよいのか考える必要があります。

現実の世界で多様性が増している一方、インターネット上では古い情報が更新されていない場合があります。例えばイギリスでは現在、ジャーナリストの47%が女性ですが、検索サイトの画像データの女性比率は33%のままだそうです。この古い情報を訓練データとしてAIが学習を進めると、偏りが再生産されてしまします。

これはインターネット上の話ですが、現実の雇用傾向も同様です。過去のデータをもとに学習するAIには、既存の社会構造を変革するような提案や判断はできません。つまりは結局のところ、私たちがどのような社会に住みたいのか、どのような社会的価値を重視するのかを、過去のデータからではなく、人間が設定する必要があるのです。

### 善い社会とはなにか



公平な判断をするAIを作るためにはデータセットやアルゴリズムを整備する必要があります。その際、何を基準にすればよいのでしょうか。この際、善い社会や善いAI を作っていく、という標語が使われる場合があります。では善い(Good)とはどのような状況なのでしょうか。誰にとって、どのような状況が善いのかなど、条件を考える必要があります。また、公平のほか、平等、公正などいくつか基準となる考え方がありますが、これらの言葉の意味は多様です。たとえば平等に関しても、スタートラインが同じであるという「**機会の平等**」と、最終的にゴールしたタイミングが同じという「**結果の平等**」は異なります。このように、何を求めるかが異なると、選択するデータやアルゴリズムも変わってきます。

#### 平等と公平


**平等(Equility)**と**公平(Equity)**も異なります。平等とは、個人的な能力や資源、経験などそれぞれの条件や状況を考慮に入れずに、全員に対して一律に同じような支援を行うことです。その結果、例えば、毎月10万円の支援で子供を高校に行かせられる家庭と、それでは不十分で高校に通わせることのできない家庭があるように、恩恵にあずかることのできる人とできない人が生じます。一方、公平とは、個人の状況や文脈に応じて支援を行うことです。その結果、全員が結果的に同じ恩恵を受けることができます。しかし、特定の人だけがより多い支援を受ける場合があるので、それが他の人から見たら「不平等」と思われる可能性もあります。現在、格差が拡大されているとされているのは、特別な支援を必要としている人ではなく、すでに恩恵を受けている人が、ますます恩恵を受けるような社会構造になっているからと指摘されています。


## **アルゴリズムの公平性や倫理的なAI開発へ**



データを集めたり、アルゴリズムを選択する基準は、どのような価値や社会を実現したいかに準じる必要があります。また、機会の平等や結果の平等を実現したいと思っても、データが少なくて難しかったり、コストがかかりすぎて対応できないといった現実的な問題もあるでしょう。特に、企業などで、あまりコストをかけないで開発したい場合は、あえて時間やお金をかけてデータやアルゴリズムの公平性を保証しようとする意欲は少ないかもしれません。 

#### AIシステムにも求められる公平さ



しかし、AI が医療や人事採用など、さまざまな場面で用いられるようになっていることから、経済的な価値だけではなく、差別的な判断を是正して公平な判断ができるような AIシステムを、時間やお金をかけてでも開発していくべきだとする考え方が強くなってきています。これは海外ではジェンダー平等やブラックライブズマター(黒人の命も大切だ)運動などとも連動しています。現在、国連が掲げる持続可能な開発目標(SDGs)の思想とも共鳴して、AI 開発においてデータやアルゴリズムの偏りをなくすこと、そのためにも社会的な価値を多様な人たちと議論していくことの重要性が指摘されています。


#### 倫理洗浄というその場しのぎ



利用者や、研究開発に資金援助しているスポンサーなどが、AIの公平さや倫理に着目しているなか、「AIの倫理的な課題は技術だけで解決できる」とする考え方があります。 今まで紹介してきたように、訓練データや アルゴリズムを変更すれば、AI の判断は変わってきます。そこで、手元のデータやアルゴリズムを変更することで、技術的には問題は解決した、とみなすことを**倫理洗浄**(**Ethics washing**)と言います。 この洗浄という言葉はもともとはグリーン ウォッシング(Greenwashing)から来ており、本当は異なるのに、環境に配慮しているような装いやごまかしをしている(うわべだけ取り繕って奇麗にしている)ことを意味します。 データやアルゴリズムを変更して、公平や平等となるような判断をすることがなぜ、うわべけの取り繕いと言われるのでしょうか? それは洗浄をして問題が一見して解決したように見せることで、背景にある社会問題を覆い隠してしまう事例が多いからです。 
たとえば、本当に法執行機関などでAIの判断を用いていいのでしょうか?あるいは、判断が「公平」であれば、監視などの目的でAIを用いていいのでしょうか? AIは人間社会や人間の権力関係から切り離して考えることはできず、技術だけではなく、そもそもの目的やデザインも考えることが重要になります。


## **トロッコ問題**

トロッコ問題にはさまざまなバリエーションがありますが、典型的な問いは以下のようなものです。

~~~
あなたの前には電車の方向を切り替えることができるスイッチがあります。電車がこのまま直進すると線路にいる５人を轢いてしまいます。
あなたがスイッチを押して電車の方向を切り替えれば、1人を轢いてしまいます。あなたはどのような行動をとりますか。
~~~

人間ではなく老人と子供だったら、自分の知り合いだったら、など条件の違う命の重みを２つ提示され、どちらの命なら危険に晒して良いか選択を迫ります。



#### [モラルマシーンプロジェクト](https://wired.jp/2019/01/02/moral-machine/)

トロッコ問題をAIの自動運転に置き換えてMITが大規模な実験をウェブサイトで行いました。モラルマシーンプロジェクトと称される研究では、次のように問いかけます。

~~~
真っ直ぐに行くと行き止まりなので左右にハンドルを切らないといけないのですが、右に曲がると1人、左だと５人轢いてしまいます。
自動運転車はどのような判断を下すべきだと思いますか。
~~~

ウェブサイトでは２つの状況が画面に示されて、どちらか一つを選択すると、次はまた違う条件が出てきます。

MITの実験では、人数だけでなく通行人の年齢や性別、種別（犬や猫などもあります）、社会的地位などを変更して、参加した人たちの集合知から、一般的な「倫理的価値」を構築しようとしました。結果として223の国と地域の人たちから4000万件の判断を得ました。分析の結果、年齢や性別や国に関わらず、動物の命よりは人間の命、少人数よりも大人数の命を優先するという傾向は共通していた一方、他の項目に関しては欧米や日本、中南米で差があったとのことです。

#### 機械は何に基づいて判断するのか知っておく

この実験では、左右どちらにもハンドルを切らずに、自動運転車に乗車中の自分が死んでも通行人を救うという選択肢も多く選ばれたそうです。しかし、いざそのような場面になったときに「自分より通行人を救う自動運転車」に乗りたいでしょうか。この問いに対しては多くの人が「いいえ」と答えています。

このような思考実験は、人間の倫理観や人と機械の関係性について考え直す機会になります。まさにこれは**モラルマシーン**（倫理的なことを考えさせる機械）なのです。

現在、自動運転に限らず、医療診断、軍事判断、政治判断など、人間でさえ判断に迷う事例が多くあります。AIをはじめとする機会が、そのような判断のサポートをするようになる可能性は大いにあります。その際、利用するAIがどのようなデータを学習し、どのようなアルゴリズムで動き、どのような社会像を反映しているのか、私たち利用者は、少なくともその傾向を知っておく機会がなければ、正しく機械を利用することができなくなってしまいます。

#### 悪いデザインを変えるための発想の転換

私たちは、システムに「ハンドルを右に切るか、左に切るか」と問いかけられれば、必ずどちらかを選ばなければならない、と考えてしまいます。しかし、このような究極の選択を人間に迫るシステムは、思考実験としては重要であっても、現実世界では「悪いデザイン」としか言いようがありません。

現実的に私たちがやるべきことは、「右か左か」という苦渋の選択を強いることではなく、「そのような判断するのかを迫られる状況に陥らないためには、どのように機械を設計すれば良いか」というように発想を転換することです。

# **3. AIプロジェクト**

AIプロジェクトの進め方として重要なことは、これまでのシステムインテグレーター的なシステム開発（ウォーターフォール）とは全く異なる点です。人が設計し、人の想定通りに稼働する従来のシステムとは異なり、AIに基づいたシステムは、全ての想定を織り込むことは原理的に不可能になります。その前提を理解した上で全体像を把握し、アジャイル的なプロジェクト•マネジメントを行わなければなりません。

実際にAIそのものを開発するフェーズは、AIプロジェクト遂行における一部でしかなく、前後に膨大なプロセスが存在します。さらに、実はその前後がAIプロジェクトを成功させる上で重要なプロセスになります。そうしたプロセスはいくつかのフレームワークや概念（**CRISP-DM**, **MLOps**）として体系化されています。



*   CRISP-DMとは、CRoss-Industry Standard Process for Data Miningの略で、データマイニングのための産業横断型標準プロセスのことです。６つのプロセス（1.ビジネスの理解、2.データの理解、3.データの準備、4.モデリング、5.評価、6.展開）からなります。この６つのステップは１から６の順に完全に行われるわけではなく、相互に行ったり来たりするプロセスが発生することになります。
*   MLOpsとは、Machine LearningとOperationsを合成した造語であり、AIを本番環境で開発しながら運用するまでの概念です。全てのプロセスを1回だけやれば良いということではなく、システム運用開始後も継続してプロセスを回すことで、より推論精度が高く安定したシステムにすることが重要です。

## **AIプロジェクトの進め方**

AIシステムは不確かな状況でのプロジェクト推進になるため、そもそも何を作りたいのかと、実現方法の各論点を事前に明確化することが重要です。


### AIを適用するべきか検討する



*   そもそもAIを適用する必要があるのかを考察し、利活用した場合の利益計画を立てて投資判断を行う。AIは目的でなく、手段の一つでしかないため、AIの特性を理解した上で判断する必要があります。例えば、ディープラーニングを使うべきなのか、ルールベースを使うべきなのかは、適用場面に応じて判断が分かれます。
*   ビジネス・技術上、組み込むべきデータのフィードバックの機構をどのようなものにするかを検討する。データが蓄積され、データのフィードバックが行われ、そのデータをAIが継続的に学習することで、推論精度が向上し、より少ないコストで大きな成果を生み出すことができるようになります。



### AIを適用した場合のプロセスを再設計する

AIを利活用する場合は**BPR**(Business Process Re-engineering)が発生します。現業務プロセスがアナログ空間にいる「人」が実行することを前提としたプロセスである場合、AIを利活用するプロセスに変更する必要があるためです。まずは、現プロセスのコストとAI導入後のコストを算出し、それぞれが乖離（特にAI導入後のコストが増大する）する場合は、AIの適用箇所と技術の連携範囲を再検討する必要があります。

### AIシステムの提供方法を決める

従来、「納品」する形式が広く取られていました。しかし、この形式はAIシステムの提供には向きません。AIを用いたシステムを開発する場合には、サービスとして提供し、運用することが望ましいのです。提供形態としては、クラウド上でWebサービスとして提供する方法や、エッジデバイスにモデルをダウンロードし、常に最新の状態となるように更新を続けるという方式があります。

### 開発計画を策定する

AIの開発は、データを確認する段階、モデルを試作する段階、運用に向けた開発をする段階にフェーズを細かく区切り、収集されたデータの中身や学習して得られたモデルの精度に応じて、柔軟に方針を修正できるような体制が望ましいです。なお、こうした特徴と開発計画に適した契約形式を採用し、適時適切にコミュニケーションをとって、プロジェクト管理をすることも必要になります。

### プロジェクト体制を構築する

AIシステムを開発するためには、開発段階から多岐にわたるスキルを保有する様々なステークホルダーを巻き込んだ体制作りが必要となります。

AIプロジェクトを推進するために必要とするスキルの範囲は非常に広く、AI関する研究開発者、それを支えるシステムを構築する技術者だけでなく、ビジネス的な観点も踏まえて全体を把握・意思決定するマネージャや、システムを使う際のUI(User Interface)/UX(User eXperience)に関するデザイナー、AIモデルを開発するデータサイエンティストを含むトータルなチーム構成が重要となります。また法的、倫理的な課題を検討するためには開発段階からの経営者、法務・経営企画・広報間の連携も重要となります。
